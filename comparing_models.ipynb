```python
#imports
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.linear_model import LinearRegression 
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import Ridge, Lasso, ElasticNet
```

```python
# get the dataframes
categorical_df=pd.read_csv("categorical.csv")
numerical_df=pd.read_csv("numerical.csv")
```

```python
categorical_df
```

```python
# concat the 2 dataframes
data =pd.concat([categorical_df,numerical_df],axis=1)
data
```

```python
data.isna().sum()
```

```python
data=data.dropna()
data
```

```python
X=data.drop("total_claim_amount",axis=1)
y=data["total_claim_amount"]
```

```python
# training test splits
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#normalize categorical data
X_train_num = X_train.select_dtypes(include = np.number)

# Scaling data
transformer = MinMaxScaler().fit(X_train_num) # need to keep transformer
X_train_normalized = transformer.transform(X_train_num)
X_train_norm = pd.DataFrame(X_train_normalized)
X_train_norm
```

### Encode categoricals


```python
#i will first split them into nominal and ordinal
X_train_cat_ordinal = X_train[["coverage","education","vehicle_size","policy_level"]]
X_train_cat_nominal = X_train[['state', 'response',"employmentstatus", 'gender', 'location_code','marital_status', 'policy_type', 'renew_offer_type', 'sales_channel','vehicle_class']]

X_train_cat_ordinal
```

```python
# one hot encode nominals 
nominal_encoder = OneHotEncoder(drop="first").fit(X_train_cat_nominal)
encoded_nominals=nominal_encoder.transform(X_train_cat_nominal).toarray()
encoded_nominals
```

```python
#put the ordinals in order
coverage_order = ['Basic', 'Extended', 'Premium']
education_order = ['High School or Below', 'College', 'Bachelor', 'Master', 'Doctor']
vehicle_size_order = ['Small', 'Medsize', 'Large']
policy_level_order = [1, 2, 3]  
```

```python
# creating the encoder for the ordinal encoding with the categories
ordinal_encoder = OrdinalEncoder(categories=[coverage_order, education_order, vehicle_size_order, policy_level_order])
```

```python
# apply it 
encoded_ordinals= ordinal_encoder.fit_transform(X_train_cat_ordinal)
encoded_ordinals
```

```python
# put categoricals together again
encoded_categoricals_train = np.concatenate((encoded_ordinals, encoded_nominals), axis=1)
encoded_categoricals_train
```

```python
#put numerical and categorical TRAIN data together
X_train_transformed = np.concatenate([encoded_categoricals_train,X_train_norm],axis=1)
X_train_transformed
```

```python
X_train_trasnformed_df = pd.DataFrame(X_train_transformed)
X_train_trasnformed_df
```

### # now to transform the test data


```python
# now we are transforming our numerical TEST data
# for numericals
X_test_num = X_test.select_dtypes(include = np.number)

# Scaling data
# we use the transformer that was trained on the training data
X_test_normalized = transformer.transform(X_test_num)
X_test_norm = pd.DataFrame(X_test_normalized,columns = X_test_num.columns)
X_test_norm.head()
```

```python
# now to transform the test data
X_test_num = X_test.select_dtypes(include = np.number)

# Scaling data
# we use the transformer that was trained on the training data
X_test_normalized = transformer.transform(X_test_num)
X_test_norm = pd.DataFrame(X_test_normalized,columns = X_test_num.columns)
X_test_norm
```

```python
#encode categoricals ,again spliting ordinals and nominals
#i will first split them into nominal and ordinal
X_test_cat_ordinal = X_test[["coverage", "education", "vehicle_size", "policy_level"]]
X_test_cat_nominal = X_test[['state', 'response', "employmentstatus", 'gender', 'location_code', 'marital_status', 'policy_type', 'renew_offer_type', 'sales_channel', 'vehicle_class']]
```

```python
#nominals
encoded_nominals_test=nominal_encoder.transform(X_test_cat_nominal).toarray()
encoded_nominals_test
```

```python
#ordinals 
encoded_ordinals_test= ordinal_encoder.fit_transform(X_test_cat_ordinal)
encoded_ordinals_test
```

```python
# put categoricals together again
encoded_categoricals_test = np.concatenate((encoded_ordinals_test, encoded_nominals_test), axis=1)
encoded_categoricals_test
```

```python
# put everything togeteher 
X_test_transformed = np.concatenate([encoded_categoricals_test,X_test_norm],axis=1)
X_test_transformed
```

```python
def evaluate_regression(y_test, predictions_test):
    mse = mean_squared_error(y_test, predictions_test)
    rmse = mean_squared_error(y_test, predictions_test, squared=False)
    mae = mean_absolute_error(y_test, predictions_test)
    r2 = r2_score(y_test, predictions_test)
    
    return mse, rmse, mae, r2

mse, rmse, mae, r2 = evaluate_regression(y_test, predictions_test)
print(f"MSE: {mse}, RMSE: {rmse}, MAE: {mae}, R2: {r2}")
```

### 7.Great! Now define a function that takes a list of models and train (and tests) them so we can try a lot of them without repeating code.

```python
def test_models(X_train,X_test,y_train,y_test,models=None):
    if models is None:
        models = {
            'Linear Regression': LinearRegression(),
            'KNN': KNeighborsRegressor(),
            'MLP Regressor': MLPRegressor(max_iter=10000),
            'Random Forest': RandomForestRegressor(),
            'Gradient Boosting': GradientBoostingRegressor(),
            'SVR': SVR(),
            'Decision Tree': DecisionTreeRegressor(),
            'Ridge': Ridge(),
            'Lasso': Lasso(),
            'Elastic Net': ElasticNet()
        }
    
    results= {}
    
    for name,model in models.items():
        model.fit(X_train,y_train)    #train the model
        predictions=model.predict(X_test) #prdict on the test set
        
        mse = mean_squared_error(y_test,predictions) # Mean Squared Error
        r2 = r2_score(y_test,predictions) # R2
        
        results[name] = {"MSE":mse, "R2": r2}
        
    return results
```

```python
results = test_models(X_train_transformed,X_test_transformed,y_train,y_test)
```
